```python
import osmnx as ox
G = ox.load_graphml(osm_data_path)
osm_edges = ox.graph_to_gdfs(G, nodes=False)
osm_edges = osm_edges[osm_edges["geometry"].notnull()]
osm_edges.reset_index(drop=True, inplace=True)
osm_edges["osm_id_index"] = osm_edges.index
```

```python
crs_proj = "EPSG:32633"
osm_edges_proj = osm_edges.to_crs(crs_proj)
osm_edges_proj.info()
osm_edges_proj.head()
```
